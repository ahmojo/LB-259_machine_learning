Als Zielfeld wähle ich median_house_value. Ich möchte also den mittleren Hauswert vorhersagen.

Teilaufgabe 2.2 


1. housing_median_age → Mittelwert

```python
import csv

dateiname = "housing.csv"
werte = []

with open(dateiname, newline="", encoding="utf-8") as datei:
    reader = csv.DictReader(datei, delimiter=";")
    for zeile in reader:
        if zeile["housing_median_age"] != "":
            werte.append(float(zeile["housing_median_age"]))

summe = 0
for wert in werte:
    summe = summe + wert

mittelwert = summe / len(werte)

print("Feld: housing_median_age")
print("Mittelwert:", mittelwert)
```

2. total_rooms → Median

```python
import csv

dateiname = "housing.csv"
werte = []

with open(dateiname, newline="", encoding="utf-8") as datei:
    reader = csv.DictReader(datei, delimiter=";")
    for zeile in reader:
        if zeile["total_rooms"] != "":
            werte.append(float(zeile["total_rooms"]))

werte.sort()
n = len(werte)

if n % 2 == 1:
    median = werte[n // 2]
else:
    mitte1 = werte[n // 2 - 1]
    mitte2 = werte[n // 2]
    median = (mitte1 + mitte2) / 2

print("Feld: total_rooms")
print("Median:", median)
```

median_income → Standardabweichung

```python
import csv
import math

dateiname = "housing.csv"
werte = []

with open(dateiname, newline="", encoding="utf-8") as datei:
    reader = csv.DictReader(datei, delimiter=";")
    for zeile in reader:
        if zeile["median_income"] != "":
            werte.append(float(zeile["median_income"]))

summe = 0
for wert in werte:
    summe = summe + wert

mittelwert = summe / len(werte)

summe_quadrate = 0
for wert in werte:
    abweichung = wert - mittelwert
    summe_quadrate = summe_quadrate + abweichung * abweichung

varianz = summe_quadrate / (len(werte) - 1)
standardabweichung = math.sqrt(varianz)

print("Feld: median_income")
print("Standardabweichung:", standardabweichung)
```

Ocean_proximity → Modus

```python
import csv

dateiname = "housing.csv"
werte = []

with open(dateiname, newline="", encoding="utf-8") as datei:
    reader = csv.DictReader(datei, delimiter=";")
    for zeile in reader:
        if zeile["ocean_proximity"] != "":
            werte.append(zeile["ocean_proximity"])

haeufigkeiten = {}

for wert in werte:
    if wert in haeufigkeiten:
        haeufigkeiten[wert] = haeufigkeiten[wert] + 1
    else:
        haeufigkeiten[wert] = 1

modus = ""
max_anzahl = 0

for wert in haeufigkeiten:
    if haeufigkeiten[wert] > max_anzahl:
        max_anzahl = haeufigkeiten[wert]
        modus = wert

print("Feld: ocean_proximity")
print("Modus:", modus)
```

Teilaufgabe 2.3

Für die grafische Darstellung habe ich ein Histogramm des Feldes housing_median_age erstellt. Das Histogramm zeigt, wie sich das Alter der Häuser im Datensatz verteilt.
So lässt sich gut erkennen, in welchen Altersbereichen besonders viele Häuser liegen.


```python
import csv
import matplotlib.pyplot as plt

dateiname = "housing.csv"
werte = []

with open(dateiname, newline="", encoding="utf-8") as datei:
    reader = csv.DictReader(datei, delimiter=";")
    for zeile in reader:
        if zeile["housing_median_age"] != "":
            werte.append(float(zeile["housing_median_age"]))

plt.hist(werte, bins=20)
plt.title("Histogramm des Alters der Häuser")
plt.xlabel("Alter der Häuser")
plt.ylabel("Häufigkeit")
plt.show()
```

Teilaufgabe 2.4

Ich habe das Feld median_income mit der Min-Max-Methode skaliert. Dabei wird der kleinste Wert zu 0 und der grösste Wert zu 1. 
Alle anderen Werte liegen anschliessend zwischen 0 und 1. In der Ausgabe sind sowohl die ursprünglichen als auch einige skalierte Werte zu sehen.

```python
import csv

dateiname = "housing.csv"
werte = []

with open(dateiname, newline="", encoding="utf-8") as datei:
    reader = csv.DictReader(datei, delimiter=";")
    for zeile in reader:
        if zeile["median_income"] != "":
            werte.append(float(zeile["median_income"]))

minimum = min(werte)
maximum = max(werte)

skalierte_werte = []

for wert in werte:
    skalierter_wert = (wert - minimum) / (maximum - minimum)
    skalierte_werte.append(skalierter_wert)

print("Feld: median_income")
print("Minimum vorher:", minimum)
print("Maximum vorher:", maximum)
print("Minimum nachher:", min(skalierte_werte))
print("Maximum nachher:", max(skalierte_werte))
print()

print("Erste 5 Werte vorher und nachher:")
for i in range(5):
    print("Vorher:", werte[i], " | Nachher:", skalierte_werte[i])
```
